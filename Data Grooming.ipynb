```python
# reads each tweet text and breaks it into it's tokens and checks if in lists from above, adds 1 to column if true
docCount=0
tempTime =time.time()
gc.disable()
for row in dfShort.itertuples(index=True, name='Pandas'):
    try:
        doc = nlp(getattr(row, "text"))
        temp = doc.text.lower()
        temp = temp.split()
        #print (temp)
        
        for index, val in enumerate(temp, start=0):   # default is zero
            if any(word in val for word in clintonList):
                containsClinton= True
        for index, val in enumerate(temp, start=0):   # default is zero
            if any(word in val for word in trumpList):
                containsTrump= True
        
        if(containsClinton):
            tempDict['Clinton'].append(1)
        else:
            tempDict['Clinton'].append(0)
        if(containsTrump):
            tempDict['Trump'].append(1)
        else:
            tempDict['Trump'].append(0)
        containsClinton= False
        containsTrump= False
                
        
        #for token in doc:
        
        
    except:
        tempDict['Trump'].append(0)
        tempDict['Clinton'].append(0)
        nlp = en_core_web_sm.load()
    docCount=docCount+1
    if(docCount%25000==0):
        print(time.time()-tempTime)
        tempTime= time.time()
        print (docCount)
#dfShort.to_csv('textSubject.csv')

gc.enable()
```
